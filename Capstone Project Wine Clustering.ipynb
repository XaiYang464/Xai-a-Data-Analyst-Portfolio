```python
import numpy as np

#Reading dataset csv
data= np.genfromtxt('wine_quality.csv',delimiter=',',dtype=None, encoding='latin1',names=True)
```

```python
#reshaping variable arrays and concatenating the X variables 
x1= np.reshape(data["sulphates"],(-1,1))
x2= np.reshape(data["fixed_acidity"],(-1,1))


X= np.concatenate((x1,x2), axis=1)
```

```python
# Normalizing feature values

from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
scaler.fit(X)
X= scaler.transform(X)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:,0],X[:,1])
plt.xlabel("sulphates")
plt.ylabel("fixed_acidity")
plt.show()
```

```python
#Clustering k means = 3
from sklearn.cluster import KMeans

km_model0 = KMeans(n_clusters=3)
km_model0.fit(X)
```

```python
centroids = km_model0.cluster_centers_
clusters = km_model0.predict(X)
```

```python
for i in range(X.shape[0]):
    if clusters[i] == 0:
        plt.scatter(X[i,0],X[i,1], color='blue', s=50)
    elif clusters[i] == 1:
        plt.scatter(X[i,0],X[i,1], color='green', s=50)
    elif clusters[i] == 2:
        plt.scatter(X[i,0],X[i,1], color='yellow', s=50)

plt.scatter(centroids[:,0],centroids[:,1], marker='x', color='red', s=50)
plt.show()
```

```python
#Clustering k mean = 4
km_model1 = KMeans(n_clusters=4)
km_model1.fit(X)
centroids = km_model1.cluster_centers_
clusters = km_model1.predict(X)
```

```python
for i in range(X.shape[0]):
    if clusters[i] == 0:
        plt.scatter(X[i,0],X[i,1], color='blue', s=50)
    elif clusters[i] == 1:
        plt.scatter(X[i,0],X[i,1], color='green', s=50)
    elif clusters[i] == 2:
        plt.scatter(X[i,0],X[i,1], color='yellow', s=50)
    elif clusters[i] == 3:
        plt.scatter(X[i,0],X[i,1], color='purple', s=50)

plt.scatter(centroids[:,0],centroids[:,1], marker='x', color='red', s=50)
plt.show()
```

```python
#The better cluster for wine prediction is where k = 4
#Groups are separated and each centroid is in the cluster's center
```
